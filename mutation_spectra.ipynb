```python

import pandas as pd
import matplotlib.pyplot as plt


mut_matrix = pd.read_csv("results/point_mut_matrix.csv")
mut_matrix = mut_matrix.set_index('Unnamed: 0')
print(mut_matrix)

lst = ["A", "C", "G", "T"]


dct = dict()
for i in lst:
    for j in lst: 
        if i != j: dct[f'{i}->{j}'] = mut_matrix.at[i, j]
plt.plot(dct.keys(), dct.values(), 'o')
plt.xticks(rotation=90)

```
